```python
import pandas as pd
import numpy as np
import seaborn as snsA
import matplotlib.pyplot as plt
```

```python
pd.options.display.float_format = '{:,.2f}'.format 
def get_plot(multi_csv, sing_csv):
    dfm = pd.read_csv(multi_csv)
    dfs = pd.read_csv(sing_csv)
    acc_val_multiclass = np.diag(dfm)
    acc_val_singleclass = np.diag(dfs)

    dm = {name:val for val, name in zip(acc_val_multiclass, dfm.columns)}
    ds = {name:val for val, name in zip(acc_val_singleclass, dfs.columns)}
    df = pd.DataFrame({key:[ds[key],dm[key] ] for key in dm.keys()})
    sns.set_theme()
# Women = [115, 215, 250, 200]
# Men = [114, 230, 510, 370]

  
    n= df.shape[1]
    r = np.arange(n)
    width = 0.33
  
  
    plt.bar(r, df.iloc[0].values, color = 'b',
        width = width,
        label='Singletask')
    plt.bar(r + width, df.iloc[1].values, color = 'g',
        width = width,
        label='Multiclass')
    plt.xlabel("Class name")
    plt.ylabel("Accuracy")
# plt.title("Accuracy value")
  
# plt.grid(linestyle='--')
    plt.xticks(r + width/2,df.columns, rotation=45)
    plt.legend()
    plt.show()
    print(f"{df.mean(axis=1)*100}")

# group_df = pd.DataFrame(dict(names=names,acc_val_singleclass=acc_val_singleclass, acc_val_multiclass=acc_val_multiclass))
```

```python
get_plot("mc_cf.csv", "sc_cf.csv")
```
